## Codebeispiel 2

In diesem Codebeispiel soll einmal die Aussage des Satzes 2 empirisch überprüft werden.
Dieser besagt, dass der Schätzer für die MMD asymptotisch Normalverteilt ist unter der 
Bedingung, dass $P^{X} \neq P^{Y}$ gilt.
Dafür wird der Schätzer $h$ einmal implementiert und er wird insgesamt 1000 mal auf unterscheidliche 
Daten angwendet, eimal von der gleichen Verteilung und einmal nicht.

Doch zunächst braucht man noch eine Kernelfunktion, 
hier der Gausskernel:

```r
gauss.kernel <- function(x, y = 0, sig = 0.2){
  return(exp(-1/(2*sig**2) * abs(x - y) ** 2))
}
```

Als nächstes wird der Schätzer $h(z_{i}, z_{j}) := k(x_{i}, x_{j}) + k(y_{i}, y_{j}) - k(x_{i}, y_{j}) - k(x_{j}, y_{i})$ implementiert, wobei $z_{i} = (x_{i}, y_{i})$ (Siehe Satz 1):

```r
h <- function(k, z1, z2){
  return(k(z1[1], z2[1]) + k(z1[2], z2[2]) - k(z1[1], z2[2]) - k(z1[2], z2[1]))
}
```

Damit hat man alles für den Schätzer für die MMD 
$(\widehat{MMD}(P^{X}, P^{Y}, \mathcal{H}_{\mathcal{X}}))^{2} =
		\frac{1}{n(n-1)} \sum_{i \neq j}^{n} h(z_{i}, z_{j})$ und kann diesen nun implementieren:

```r
MMD <- function(k, z){
  n <- nrow(z)
  summe <- 0
  for(i in 1:n){
    for(j in (1:n)[-i]){
      summe <- summe + h(k, z[i,], z[j,])
    }
  }
  return(1/(n*(n-1)) * summe)
}
```

Dieser soll nun m = 1000 jeweil auf Datensätze der Größe n = 100 angewendet werden und zwar einmal für $X \sim Unif(0, 1)$ und $Y \sim Unif(1, 2)$ und einmal für $X, Y \sim Unif(0, 1)$

```r
vers1 <- replicate(m, sqrt(n) * MMD(gauss.kernel, matrix(data = c(runif(n), runif(n) + 1), ncol = 2)))
vers2 <- replicate(m, sqrt(n) * MMD(gauss.kernel, matrix(data = c(runif(n), runif(n)),ncol = 2)))
```

```r
load("Experiment.Rdata")     # Da die Berechnung serh lange dauert, hier sind vers1 und vers2 gespeichert
```

Nun kann man die empirischen Histogramme erstellen für ver1 und vers2:

```r
library('latex2exp')
```

### Versuch 1

Bei Vers1 waren die Verteilungen unterschiedlich, d.h. nach Satz 2 würde man eine Gauß-ähnliche Verteilung erwarten.

```r
hist(vers1, main = TeX("$\\widehat{MMD}^{2}$ für $P^{X} \\neq P^{Y}$"), xlab = TeX("$\\widehat{MMD}^{2}$"), ylab = "freq", freq = F)
```

In der Tat hat erkennt man hier ganz klar eine Verteilung wie bei einer Gaußverteilung, d.h. Satz 2 bestätigt sich damit.

## Versuch 2

In vers2 waren die Verteilungen gleich, d.h. hier kann man keine Aussage über die Verteilung treffen. 
Damit der Test eine gute Power hat, sollte dieser aber im besten Fall eine sehr hohe Konzentration bei niedrigen Werten haben, d.h. der größte Teil sollte klein sein. 
Das ist deswegen, da man meistens die Nullhypothese ablehnt, wenn der rauskommende Wert klein ist.

```r
hist(vers2, main = TeX("$\\widehat{MMD}^{2}$ für $P^{X} = P^{Y}$"), xlab = TeX("$\\widehat{MMD}^{2}$"), ylab = "freq", freq = F)
```

Auch hier bestätigt sich wieder dei Vermutung, denn der größte Teil nimmt sehr niedrige Werte an.
